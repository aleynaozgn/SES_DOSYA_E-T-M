```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
import os
import uuid
#MFCC GÖRÜNTÜLERİNİ KAYDETME

#  Google Drive'ı mount et (sadece bir kere yapman yeterli)
from google.colab import drive
drive.mount('/content/drive')

#  Temel yol ve etiketler
base_path = "/content/drive/MyDrive/ses_dosyaları"  # Ses dosyalarının olduğu yer
folders = {"HALK1": 0, "SANAT": 1, "POP1": 2}

#  Parametreler
SEGMENT_SECONDS = 3
SHIFT_SECONDS = 1
MAX_AUDIO_SECONDS = 9  # sadece ilk 9 saniye
IMG_SIZE = (224, 224)

# Çıktı klasörü — DRIVE içine kaydedeceğiz!!
output_dir = "/content/drive/MyDrive/mfcc_images"  # Drive içi bir klasör
os.makedirs(output_dir, exist_ok=True)

# Her klasör için işlem
for folder, label in folders.items():
    folder_path = os.path.join(base_path, folder)
    save_folder = os.path.join(output_dir, str(label))
    os.makedirs(save_folder, exist_ok=True)

    for file in os.listdir(folder_path):
        if not file.lower().endswith(".mp3"):
            continue

        file_path = os.path.join(folder_path, file)

        try:
            print(f"\n İşleniyor: {file}")
            y, sr = librosa.load(file_path, sr=None)
            print(" Uzunluk (saniye):", len(y) / sr)

            max_samples = int(MAX_AUDIO_SECONDS * sr)
            y = y[:max_samples]  # sadece ilk 9 saniyeyi al

            samples_per_segment = SEGMENT_SECONDS * sr
            shift_samples = SHIFT_SECONDS * sr

            for start in range(0, len(y) - samples_per_segment + 1, shift_samples):
                end = start + samples_per_segment
                segment = y[start:end]

                mfcc = librosa.feature.mfcc(y=segment, sr=sr, n_mfcc=13)

                # Görsel üret ve kaydet
                fig = plt.figure(figsize=(2.24, 2.24), dpi=100)
                plt.axis("off")
                librosa.display.specshow(mfcc, sr=sr)
                img_path = os.path.join(save_folder, f"{uuid.uuid4()}.png")
                plt.savefig(img_path, format="png", bbox_inches="tight", pad_inches=0)
                plt.close(fig)

                print(f" Kaydedildi: {img_path}")

        except Exception as e:
            print(f" {file} dosyası atlandı. Hata: {e}")
```
